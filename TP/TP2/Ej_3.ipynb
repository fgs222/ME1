Luego de visualizar la siguiente señal en un DSO, nos preguntamos cuál será la indicación de un voltímetro que responde a:

- TrueRMS 
- Valor pico
- Valor medio de Modulo



Considerar

- Canal vertical=5V/DIV. Punta x10.
- Ángulo de conducción de 90°

### True RMS

$$
S^2_{RMS} = \frac{1}{T}~.~\int_{0}^{T} S^2(t)~dt = \frac{2}{T}~.~\int_{\frac{\pi}{2}}^{\frac{T}{2}} V^2_p~.~sen^2(\omega t)~dt
$$

$$
S^2_{RMS} = \frac{V^2_p}{T}~.~[t - \frac{1}{4\omega}~.~sen(2\omega t)]\Bigl|_{\frac{\pi}{2}}^{\frac{T}{2}} = \frac{V^2_p}{T}~.~[\frac{T}{2} - \frac{1}{4\omega}~.~sen(2\omega~.~\frac{T}{2}) - \frac{\pi}{2} + \frac{1}{4\omega}~.~sen(2\omega~.~ \frac{\pi}{2}]
$$

$$
S^2_{RMS} = \frac{V^2_p}{T}~.~[\frac{T}{2} - \frac{1}{4\omega}~.~sen(2\omega~.~T) - \frac{\pi}{2} + \frac{1}{4\omega}~.~sen(\omega~.~\pi)]
$$

$$
S^2_{RMS} = \frac{V^2_p}{T}~.~[\frac{T}{2} - \frac{\pi}{2} - \frac{1}{4\omega}~.~sen(2\omega~.~T)]~~~~~~con~~\omega = 1~~~~T = 2\pi
$$

$$
S^2_{RMS} = \frac{V^2_p}{2\pi}~.~[\frac{2\pi}{2} - \frac{\pi}{2} - \frac{1}{4}~.~sen(4\pi)] = \frac{V^2_p}{2\pi}~.~[\pi - \frac{\pi}{2}] = \frac{V^2_p}{2\pi}~.~\frac{\pi}{2}
$$

$$
S^2_{RMS} = \frac{V^2_p}{4}~~\to~~S_{RMS} = \frac{V_p}{2}
$$

```python
from math import sqrt, pi

VP = 150

print("El valor TRUE RMS de la señal es: {} V".format(VP / 2))
```

### Valor Pico

$$
V_p = 150~V~~~~~~~~~~K_{m+} = \frac{1}{\sqrt{2}}
$$

Siendo:

$$
S_i = V_p~.~K_{m+}
$$

```python
KM = 1 / sqrt(2)

print("El valor pico de la señal es: {} V".format(VP * KM))
```

### Valor medio de modulo

$$
S_{|me|} = \frac{1}{T}~.~\int_{0}^{T} |S(t)|~dt = \frac{2}{T}~.~\int_{\frac{\pi}{2}}^{\frac{T}{2}} V_p~.~sen(\omega t)~dt
$$

$$
S_{|me|} = \frac{2V_p}{T}~.~[- cos(\omega t)]\Bigl|_{\frac{\pi}{2}}^{\frac{T}{2}} = \frac{2V_p}{T}~.~[- cos(\frac{\omega T}{2}) - cos(\frac{\omega \pi}{2})]~~~~~~con~~\omega = 1~~~~T = 2\pi
$$

$$
S_{|me|} = \frac{2V_p}{2\pi}~.~[- cos(\frac{2\pi}{2}) - cos(\frac{\pi}{2})] = \frac{V_p}{\pi}~.~[- cos(\pi) - cos(\frac{\pi}{2})]
$$

$$
S_{|me|} = \frac{V_p}{\pi}
$$

Siendo:

$$
S_i = S_{me}~.~K_{|me|}~~~~~~~con~~k_{|me|} = \frac{\pi}{2\sqrt{2}}
$$

```python
KME = pi / (2 * sqrt(2))
SME = VP / pi

print("El valor medio de la señal es: {} V".format(SME * KME))
```
