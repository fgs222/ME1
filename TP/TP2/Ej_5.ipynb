### Parte A

Siendo el factor de forma para una señal senoidal:

$$
F_{RMS,P} = \frac{1}{\sqrt{2}}
$$

Siendo el valor eficaz de la señal:

$$
S_e = F_{RMS,P}~.~V_p
$$

```python
from math import sqrt, pi

F_RMS_P = 1 / sqrt(2)
V_P = 18

s_e = F_RMS_P * V_P

print("El valor eficaz de la señal es: {:.2f} V".format(s_e))
```

### Parte B

Siendo la señal de valor medio de modulo:

$$
S_{|me|} = \frac{1}{T}~.~\int_{0}^{T} |S(t)|~.~dt = \frac{2}{T}~.~\int_{0}^{\frac{T}{2}} V_p~.~sin(\omega t)~.~dt = \frac{2V_p}{T}~.~\int_{0}^{\frac{T}{2}} sin(\omega t)~.~dt
$$

$$
S_{|me|} = \frac{2V_p}{T}~.~(- \frac{cos(\omega t)}{\omega})\Bigl|_{0}^{\frac{T}{2}} = \frac{2V_p}{2\pi}~.~2~~\to~~S_{|me|} = \frac{2V_p}{\pi}
$$

```python
s_me = 2 * V_P / pi

print("El valor indicado por el instrumento es: {:.2f} V".format(s_me))
```

### Parte C

Siendo la constante de calibracion para valor medio de modulo:

$$
K_{|me|} = \frac{\pi}{2\sqrt{2}}
$$

Siendo la indicacion de un instrumento que responde a valor RMS de AC:

$$
S_e = S_{|me|}~.~K_{|me|} 
$$

$$
S_{RMS}(t) = \frac{1}{T}~.~\int_{0}^{T} S^2(t)~.~dt = \frac{2}{T}~.~V^2_p~.~\int_{0}^{\frac{T}{2}} sin^2(\omega t)~.~dt - V^2_{med} \int_{0}^{\frac{T}{2}} dt
$$

$$
S_{RMS}(t) = V^2_p~.~(\frac{t}{2} - \frac{sen(2t\omega)}{4\omega})\Bigl|_{0}^{\frac{T}{2}} - V^2_{med}~.~\frac{T}{2} = V^2_p~.~(\frac{T}{4} - \frac{sen(T\omega)}{4\omega}) - V^2_{med}~.~\frac{T}{2}
$$

$$
S_{RMS}(t) = V^2_p~.~(\frac{T}{4} - \frac{sen(T\omega)}{4\omega}) - V^2_{med}~.~\frac{T}{2}
$$

```python
K_ME = pi / (2 * sqrt(2))

s_e = s_me * K_ME

print("El valor indicado por el instrumento es: {:.2f} V".format(s_e))
```
