### Parte A

Siendo la tension pico a pico del generador:

$$
V_{pp}(g) = n_{div}~.~V_{div}
$$

Siendo la constante del valor eficaz para un valor pico a pico:

$$
K_{RMS} = \frac{1}{2\sqrt{2}}
$$

Siendo la potencia activa del generador:

$$
P(g) = \frac{ (V_{pp}~.~K_{RMS})^2 }{R}
$$

```python
from math import sqrt

N_DIV = 5.2
V_DIV = 1
R = 100

vpp = N_DIV * V_DIV
K_RMS = 1 / ( 2 * sqrt(2) )
p_g = (vpp * K_RMS) ** 2 / R

print("El valor de la tension pico a pico es: {} volts".format(vpp))
print("El valor de la constante k es: {}".format(K_RMS))
print("El valor de la potencia activa del generador es: {} W".format(p_g))
```

Siendo el error tipo B del canal 2 del osciloscopio:

$$
u_j(osc) = \frac{a}{\sqrt{3}}~~~~~~~con~~~a = 0,03~.~V_{pp} + 0,05~div~.~1~\frac{V}{div}
$$

Siendo el cuadrado de la incertidumbre combinada de la tension pico a pico:

$$
u^2_c(V_{pp}) = u^2_j(osc) + err^2_{lectura}~~~~~~~~~~~~err_{lectura} = \frac{0,2~div~.~V_{div}}{\sqrt{3}}
$$

```python
err_lect =  0.2 * V_DIV / sqrt(3)
a = 0.03 * vpp + 0.05 * V_DIV    # a = 3% * reading + 0.05 div
uj_osc = a / sqrt(3)
uc_vpp = sqrt(uj_osc ** 2 + err_lect ** 2)

print("El valor de la precision es: {} V".format(a))
print("El error tipo B del canal 2 del osciloscopio es: {} V".format(uj_osc))
print("El valor de la incertidumbre combinada de Vpp es: {} V".format(uc_vpp))
```

Siendo la incertidumbre al cuadrado de la potencia del generador:

$$
u^2_c(p_g) = (\frac{2~.~V_{pp}~.~K^2_{RMS}}{R})^2~.~u^2_c(V_{pp}) + (\frac{-(V_{pp}~.~K_{RMS})^2}{R^2})^2~.~u^2_c(R)
$$

Dado que la incertidumbre de la resistencia es nula:

$$
u^2_c(p_g) = (\frac{2~.~V_{pp}~.~K^2_{RMS}}{R})^2~.~u^2_c(V_{pp})
$$

```python
d = (2 * vpp * K_RMS ** 2) / R
uc_g = sqrt( d ** 2 * uc_vpp ** 2 )
k = 2
u_g = k * uc_g

print("La incertidumbre combinada de la potencia activa del generador es: {} W".format(uc_g))
print("La incertidumbre expandida de la potencia activa del generador es: {} W".format(u_g))
```

Expandiendo para un factor de cobertura del 95,45%, el valor de la potencia activa del generador es:

$$
P_g = (33,80~;\pm~4,31)~mW
$$

### Parte B

Siendo la tension pico a pico del generador:

$$
V_{pp}(g) = n_{div}~.~V_{div}
$$

Siendo la constante del valor eficaz para un valor pico a pico:

$$
K_{RMS} = \frac{1}{2\sqrt{2}}
$$

Siendo la corriente eficaz del circuito:

$$
I_{RMS} = \frac{V_{pp}~.~K_{RMS}}{R}
$$

Siendo la potencia aparente:

$$
S_g = V_{pp}~.~K_{RMS}~.~I_{RMS}
$$

```python
N_DIV = 10

vpp_2 = N_DIV * V_DIV
i_rms = vpp * K_RMS / R
s_g = vpp_2 * K_RMS * i_rms

print("El valor de la tension pico a pico es: {} V".format(vpp_2))
print("El valor de la corriente eficaz es: {} A".format(i_rms))
print("El valor de la potencia aparente es: {} VA".format(s_g))
```

Siendo el error tipo B del canal 1 osciloscopio:

$$
u_j(osc) = \frac{a}{\sqrt{3}}~~~~~~~con~~~a = 0,03~.~V_{pp} + 0,05~div~.~1~\frac{V}{div}
$$

Siendo el cuadrado de la incertidumbre combinada de la tension pico a pico:

$$
u^2_c(V_{pp}) = u^2_j(osc) + err^2_{lectura}~~~~~~~~~~~~err_{lectura} = \frac{0,2~div~.~V_{div}}{\sqrt{3}}
$$

```python
a1 = 0.03 * vpp + 0.05 * V_DIV
uj_osc1 = a / sqrt(3)
uc_vpp1 = sqrt(uj_osc1 ** 2 + err_lect ** 2)

print("El valor de la precision es: {} V".format(a1))
print("El error tipo B del canal 2 del osciloscopio es: {} volts".format(uj_osc1))
print("El valor de la incertidumbre combinada de Vpp es: {} V".format(uc_vpp1))
```

Siendo la incertidumbre combinada de la potencia aparente:

$$
u_c(S_g) = (\frac{2~.~V_{pp}~.~K^2_{RMS}}{R})^2~.~u^2_c(V_{pp}) + (\frac{-(V_{pp}~.~K_{RMS})^2}{R^2})^2~.~u^2_c(R)
$$

Dado que la incertidumbre de la resistencia es nula:

$$
u^2_c(p_g) = (\frac{2~.~V_{pp}~.~K^2_{RMS}}{R})^2~.~u^2_c(V_{pp})
$$

```python
d2 = (2 * vpp_2 * K_RMS ** 2) / R
uc_s = sqrt( d2 ** 2 * uc_vpp1 ** 2 )
k = 2
u_s = k * uc_s

print("La incertidumbre combinada de la potencia aparente del generador es: {} VA".format(uc_s))
print("La incertidumbre expandida de la potencia aparente del generador es: {} VA".format(u_s))
```

Expandiendo para un factor de cobertura del 95,45%, el valor de la potencia aparente del generador es:

$$
S_g = (65,00~;\pm~8,29)~mVA
$$

### Parte C

Siendo la potencia reactiva:

$$
Q_g = \sqrt{S^2_g - P^2_g}
$$

Siendo la impedancia capacitiva:

$$
Z_c = \frac{Q_g}{I^2_{RMS}}
$$

Siendo la velocidad angular:

$$
\omega = \frac{2\pi}{T}
$$

Siendo la capacidad:

$$
C = \frac{1}{Z_c~.~\omega}
$$

```python
from math import pi

T = 100e-6 * N_DIV

q_g = sqrt( s_g ** 2 - p_g ** 2 )
zc = q_g / i_rms ** 2
omega = 2 * pi / T
C = 1 / (zc * omega)

print("El valor de la potencia reactiva es: {} VAr".format(q_g))
print("El valor de la impedancia capacitiva es: {} ohms".format(zc))
print("El valor de la velocidad angular es: {} 1/s".format(omega))
print("El valor de la capacidad es: {} F".format(C))
```

Por lo tanto, el valor del capacitor es:

$$
C = 968,90~nF
$$
