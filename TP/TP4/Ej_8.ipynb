Siendo el tiempo indicado en la pantalla:

$$
t_i = \sqrt{t^2_g + 2~.~t^2_{coaxil} + t^2_{eo} + t^2_{dbe}}
$$

Despejando para el tiempo de establecimiento del dispositivo bajo ensayo:

$$
t_{dbe} = \sqrt{t^2_i - t^2_g - 2~.~t^2_{coaxil} - t^2_{eo}}
$$

Siendo el tiempo de establecimiento del osciloscopio:

$$
t_{eo} = \frac{0,35}{B_w}
$$

```python
from math import sqrt

TI = 1e-3
TG = 10e-9
TC = 0.1e-9
BW = 80e6

teo = 0.35 / BW
tdbe = sqrt(TI ** 2 - TG ** 2 - 2 * TC ** 2 - teo ** 2)

print("El tiempo de establecimiento del dispositivo bajo ensayo es: {} s".format(tdbe))
```

Por lo tanto, el tiempo de establecimiento del dispositivo bajo ensayo es:

$$
t_{dbe} = 0,99 ms
$$
