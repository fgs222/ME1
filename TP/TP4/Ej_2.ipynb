Siendo la constante del valor eficaz para un valor pico:

$$
K_{RMS} = \frac{1}{\sqrt{2}}
$$

Siendo la potencia aparente del generador:

$$
S_g = V_p~.~K_{RMS}~.~I_{RMS}
$$

```python
from math import sqrt

VP = 5
I_RMS = 27.66e-3
K_RMS = 1 / sqrt(2)

s_g = VP * K_RMS * I_RMS

print("El valor de la potencia aparente del generador es: {} VA".format(s_g))
```

Siendo la potencia activa pico a pico del osciloscopio:

$$
P_{pp}(g) = V^+_p - V^-_p = 195~mW
$$

Siendo la constante del valor eficaz para un valor pico a pico:

$$
K_{RMS} = \frac{1}{2\sqrt{2}}
$$

Siendo la potencia activa eficaz:

$$
P(g) = P_{pp}(g)~.~K_{RMS}
$$

Siendo el valor de la resistencia:

$$
R = \frac{P(g)}{I^2_{RMS}}
$$

```python
PPP_G = 195e-3
K_RMS = 1 / (2 * sqrt(2) )

p_g = PPP_G * K_RMS
r = p_g / I_RMS ** 2

print("El valor de la potencia aparente del generador es: {} W".format(p_g))
print("El valor de la resistencia es: {} ohms".format(r))
```

Por lo tanto, el valor de la resistencia es:

$$
R = 90,11~\Omega
$$

Siendo la potencia reactiva:

$$
Q_g = \sqrt{S^2_g - P^2_g}
$$

Siendo la impedancia capacitiva:

$$
Z_c = \frac{Q_g}{I^2_{RMS}}
$$

Debido a que la frecuencia de la potencia es el doble de la frecuencia de la tension, la velocidad angular de la tension es:

$$
\omega = \frac{\pi}{T}
$$

Siendo la capacidad:

$$
C = \frac{1}{Z_c~.~\omega}
$$

```python
from math import pi

T = 0.5

q_g = sqrt( s_g ** 2 - p_g ** 2 )
zc = q_g / I_RMS ** 2
omega = pi / T
C = 1 / (zc * omega)

print("El valor de la potencia reactiva es: {} VAr".format(q_g))
print("El valor de la impedancia capacitiva es: {} ohms".format(zc))
print("El valor de la velocidad angular es: {} 1/s".format(omega))
print("El valor de la capacidad es: {} F".format(C))
```

Por lo tanto, el valor del capacitor es:

$$
C = 1755,65~\mu F
$$
