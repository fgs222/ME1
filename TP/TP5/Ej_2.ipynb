# Ejercicio 2

### Parte A

Siendo la estabilidad de la base de tiempos:

$$
|\frac{\Delta f_{BT}}{f_{BT}}| = |\frac{1~Hz}{10~MHz}|~~\to~~|\frac{\Delta f_{BT}}{f_{BT}}| = 1 \times 10^{-7} = 0,1~ppm
$$

Siendo el numero de cuentas indicadas:

$$
N = \frac{T_c}{T_x} = \frac{10~s}{\frac{1}{10~kHz}}~~\to~~N = 1 \times 10^5
$$

Siendo la incertidumbre de cuantificacion:

$$
|\frac{\Delta N}{N}| = \frac{1}{1 \times 10^5}~~\to~~|\frac{\Delta N}{N}| = 1 \times 10^{-5} = 10~ppm
$$

Siendo la ecuacion para el calculo de la incertidumbre en la medicion de frecuencia:

$$
\frac{\Delta f_I}{f_I} = \pm~( |\frac{\Delta f_{BT}}{f_{BT}}| + |\frac{\Delta N}{N}| ) = \pm~(0,1~ppm + 10~ppm)~~\to~~\frac{\Delta f_I}{f_I} = 10,1~ppm
$$

Despejando la incertidumbre de frecuencia:

$$
\Delta f_I = 10,1~ppm~.~f_I = 10,1 \times 10^{-6}~.~10~kHz~~\to~~\Delta f_I = 0,101~Hz
$$

Desnormalizando la incertidumbre de la frecuencia, se obtiene la incertidumbre relativa:

$$
u(f_I) = \frac{\Delta f_I}{\sqrt{3}} = \frac{0,101~Hz}{\sqrt{3}}~~\to~~u(\Delta f_I) = 0,05831237719~Hz
$$

Expandiendo el valor de la incertidumbre para una distribucion del tipo T de student (ya que no se cuenta con error tipo A) para un intervalo de confianza del 95,45%,:

$$
U(f_I) = k~.~u(f_I) = 1,65~.~0,05831237719~Hz~~\to~~U(f_I) = 0,09621542236~Hz
$$

Por lo tanto el valor de la frecuencia incognita es:

$$
f_I = (10\times10^3~;\pm~0,096)~Hz
$$

### Parte B

Siendo la estabilidad de la base de los tiempos:

$$
|\frac{\Delta f_{BT}}{f_{BT}}| = |\frac{1~Hz}{10~MHz}|~~\to~~|\frac{\Delta f_{BT}}{f_{BT}}| = 1 \times 10^{-7} = 0,1~ppm
$$

Siendo el numero de cuentas indicadas:

$$
N = \frac{10^M~.~T_x}{T_c} = \frac{10^1~.~\frac{1}{10~kHz}}{\frac{1}{10~MHz}}~~\to~~N = 1\times10^4
$$

Siendo la incertidumbre de cuantificacion:

$$
|\frac{\Delta N}{N}| = \frac{1}{1 \times 10^4}~~\to~~|\frac{\Delta N}{N}| = 1 \times 10^{-4} = 100~ppm
$$

Siendo el error de disparo:

$$
e_D = 0,5\% ~~\to~~e_D = 0,005 = 5000~ppm
$$

Con los valores calculados anteriormente, se procede a calcular la incertidumbre:

$$
\frac{\Delta T_I}{T_I} = \pm~( |\frac{\Delta f_{BT}}{f_{BT}}| + |\frac{\Delta N}{N}| + |e_D|) = \pm~(0,1~ppm + 100~ppm + 5000~ppm)~~\to~~\frac{\Delta T_I}{T_I} = \pm~5100,1~ppm
$$

Desnormalizando la incertidumbre relativa de la frecuencia, se obtiene la incertidumbre relativa:

$$
u_R(T_I) = \frac{\Delta T_I}{\sqrt{3}~.~T_I} = \frac{5,1001\times10^{-7}~s}{\sqrt{3}~.~10\times10^{-3}~s}~~\to~~u_R(T_I) = 2,944544108\times10^{-5}
$$

Sabiendo que el periodo es la inversa de la frecuencia y aplicando el caso particular para el calculo de la incertidumbre:

$$
y = C~.~x^{E_1}~.~x^{E_2}~.~~.~x^{E_3}~~\to~~T = \frac{1}{f}
$$

Por lo tanto, la incertidumbre combinada relativa del periodo incognita es:

$$
u_{cR}(T_I) = \sqrt{[E_1~.~u_R(f_I)]^2} = \sqrt{[-1~.~u_R(f_I)]^2}
$$

Se observa que:

$$
u_{cR}(T_I) = u_{cR}(f_I)
$$

Expandiendo el valor de la incertidumbre para un intervalo de confianza del 95,45%,:

$$
U(f_I) = k~.~u_{cR}(T_I)~.~f_I = 2~.~2,944544108\times10^{-5}~.~10~kHz~~\to~~U(f_I) = 0,5889088216~Hz
$$

Por lo tanto el valor del periodo incognita es:

$$
T_I = (10\times10^3~;\pm0,589)~Hz
$$

La incertidumbre con un medidor de periodo es mucho mayor que la que se obtiene con un medidor de frecuencia. Esto se debe a que, en el caso del medidor de periodo, se tiene un error de disparo que en este caso es bastante grande.
