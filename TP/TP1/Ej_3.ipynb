# Ejercicio 3

Se busca calibrar la función amperímetro de alterna de un multímetro digital de 3 ½ dígitos. Se controlará en este ejercicio sólo el punto de fondo de escala de 10A, con 50 Hz. Se empleará un calibrador (aparato que provee la corriente necesaria y la indicación de su valor, 10A en este caso, con alta exactitud).



Se toman 5 mediciones sucesivas en el instrumento a contrastar que arrojan los siguientes valores:

| N     | 1     | 2     | 3     | 4     | 5     |
|-------|-------|-------|-------|-------|-------|
| I [A] | 10,01 | 10,00 | 10,02 | 10,01 | 10,00 |

El fabricante del calibrador específica en su catálogo que la incertidumbre expandida de este dispositivo es ±(0,05% lectura + 2 mA) con una probabilidad de 99% y distribución gaussiana. Estime el error y la incertidumbre expandida en el error de la medida de 10 A, con un factor de cobertura del 95 %.


Siendo la media aritmetica:

$$
\bar{X} = \frac{1}{N}\sum_{k=1}^{N} X_k
$$

Calculamos la media aritmetica de I:

$$
\bar{I} = \frac{10,01+10,00+10,02+10,01+10,00}{5}~A~~\to~~\bar{I} = 10,008 A
$$

Siendo la desviacion estandar:

$$
s(X) = \sqrt{\frac{1}{N-1}\sum_{k=1}^{N} (X_k - \bar{X})^2}
$$

Calculamos la desviacion estandar de I:

$$
s(I) = \sqrt{\frac{1}{5-1}\sum_{k=1}^{N} (X_k - 10,008~A)^2}
$$

Con:

$$
\sum_{k=1}^{N} (X_k - 10,008~A)^2 = (10,00~A - 10,008~A)^2 + (10,01~A - 10,008~A)^2 + (10,02~A - 10,008~A)^2 + (10,01~A - 10,008~A)^2 + (10,00~A - 10,008~A)^2
$$

$$
\sum_{k=1}^{N} (X_k - 10,008~A)^2 = 2,8\times10^{-4}~A^2
$$

$$
s(I) = \sqrt{\frac{2,8\times10^{-4}~A^2}{4}}~~\to~~s(I) = 8,366\times10^{-3}~A
$$

Siendo el error tipo A del amperimetro:

$$
u_i = \frac{S}{\sqrt{N}}~~~~~~~S: desviacion~estandard~~~~N: cantidad~de~mediciones
$$

Calculamos el error tipo A para I:

$$
u_i(amp) = \frac{8,366\times10^{-3}~A}{\sqrt{5}}~~\to~~u_i(amp) = 3,74\times10^{-3}~A
$$

Siendo el error tipo B del amperimetro:

$$
u_j(amp) = \frac{\frac{resolucion}{2}}{\sqrt{3}}~~~~~~siendo~la~resolucion~el~error~por~redondeo~del~instrumento
$$

$$
u_j(amp) = \frac{\frac{0,001~A}{2}}{\sqrt{3}}~~\to~~u_j(amp) = 0,288675\times10^{-6}~A
$$

Siendo el error tipo B del instrumento patron:

$$
u_j(cal) = \sqrt{Var(x)}~~~~~~~Var(x):~varianza = \int_{a}^b x^2f(x)dx
$$

Asumiendo que la medicion tiene una distribucion uniforme continua:

$$
Var(x) = \frac{a^2}{3}~~~~~~a: Precision
$$

Siendo la precision:

$$
a = \Delta V .err\%~~~~~~\Delta V: valor~de~medicion~~~~err\%: error~total~porcentual
$$

Siendo el error porcentual  total:

$$
err\% = err_{med}\% + err_{cont}\%~~~~~~err_{med}\%: error~de~medicion~~~~err_{cont}\%: error~de~conteo
$$

Donde el error porcentual de conteo es:

$$
err_{cont}\% = \frac{total~de~conteos}{valor~de~cuentas~del~instrumento}~.~100\%
$$

$$
err_{cont}(2~conteos)\% = \frac{2}{1000}~.100\%~~\to~~err_{cont}(2~conteos)\% = 0,2\%
$$

$$
err\% = 0,05\% + 0,2\%~~\to~~err\% = 0,25\%
$$

$$
a = \pm~10~A~.~0,25\%~~\to~~a = \pm~0,025~A
$$

Calculando el error tipo B del instrumento patron:

$$
u_j(cal) = \sqrt{\frac{a^2}{3}} = \frac{a}{\sqrt{3}}~~\to~~u_j(cal) = \frac{0,025~A}{\sqrt{3}}~~\to~~u_j(cal) = 0,0144~A
$$

Calculando la incertidumbre combinada del amperimetro:

$$
u_c(amp) = \sqrt{u_i^2+u_j^2}~~\to~~u_c(amp) = \sqrt{(3,74\times10^{-3}~A)^2 + (0,288675\times10^{-6}~A)^2}~~\to~~u_c(amp) = 3,74\times10^{-3}~A
$$

Siendo la incertidumbre combinada del instrumento patron:

$$
u_c(cal) = u_j(cal)~~\to~~u_c(cal) = 0,0144~A
$$

Se calcula la incertidumbre combinada del valor medido y el valor verdadero:

$$
u_c(e_{cal}) = \sqrt{u^2_c(amp) + u^2_c(cal)}~~\to~~u_c(e_{cal}) = \sqrt{(3,74\times10^{-3}~A)^2 + (0,0144~A)^2}~~\to~~u_c(e_{cal}) = 0,0148777521~A
$$

Dado que la incertidumbre combinada es la ocnvolucion de una distribucion normal por una distribucion uniforme:

$$
asumimos~distribucion~normal \therefore k = 2
$$

Siendo la incertidumbre expandida:

$$
U(e_{cal}) = k_{expansion} . u_c(e_{cal}) = 2~.~0,0148777521~A~~\to~~U(e_{cal}) = 0,02975551041~A
$$

Siendo el error de calibracion:

$$
Error = |I_{med} - I_{real}| = |10,008~A - 10~A|~~\to~~Error = 0,008~A
$$

Por lo que el valor del error de calibracion con un grado de confianza de 95% es:

$$
\hat{e_{cal}} = (0,008~;\pm~0,030)~A
$$
