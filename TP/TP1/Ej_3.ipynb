# Ejercicio 3

Se busca calibrar la funcin ampermetro de alterna de un multmetro digital de 3  dgitos. Se controlar en este ejercicio slo el punto de fondo de escala de 10A, con 50 Hz. Se emplear un calibrador (aparato que provee la corriente necesaria y la indicacin de su valor, 10A en este caso, con alta exactitud).



Se toman 5 mediciones sucesivas en el instrumento a contrastar que arrojan los siguientes valores:

| N     | 1     | 2     | 3     | 4     | 5     |
|-------|-------|-------|-------|-------|-------|
| I [A] | 10,01 | 10,00 | 10,02 | 10,01 | 10,00 |

El fabricante del calibrador especfica en su catlogo que la incertidumbre expandida de este dispositivo es (0,05% lectura + 2 mA) con una probabilidad de 99% y distribucin gaussiana. Estime el error y la incertidumbre expandida en el error de la medida de 10 A, con un factor de cobertura del 95 %.


Siendo la media aritmetica:

$$
\bar{X} = \frac{1}{N}\sum_{k=1}^{N} X_k
$$

Calculamos la media aritmetica de I:

$$
\bar{I} = \frac{10,01+10,00+10,02+10,01+10,00}{5}~A~~\to~~\bar{I} = 10,008 A
$$

Siendo la desviacion estandar:

$$
s(X) = \sqrt{\frac{1}{N-1}\sum_{k=1}^{N} (X_k - \bar{X})^2}
$$

Calculamos la desviacion estandar de I:

$$
s(I) = \sqrt{\frac{1}{5-1}\sum_{k=1}^{N} (X_k - 10,008~A)^2}
$$

Con:

$$
\sum_{k=1}^{N} (X_k - 10,008~A)^2 = (10,00~A - 10,008~A)^2 + (10,01~A - 10,008~A)^2 + (10,02~A - 10,008~A)^2 + (10,01~A - 10,008~A)^2 + (10,00~A - 10,008~A)^2
$$

$$
\sum_{k=1}^{N} (X_k - 10,008~A)^2 = 2,8x10^{-4}~A^2
$$

$$
s(I) = \sqrt{\frac{2,8x10^{-4}~A^2}{4}}~~\to~~s(I) = 8,366x10^{-3}~A
$$

Siendo el error tipo A:

$$
u_i = \frac{S}{\sqrt{N}}~~~~~~~S: desviacion~estandard~~~~N: cantidad~de~mediciones
$$

Calculamos el error tipo A para I:

$$
u_i(I) = \frac{8,366x10^{-3}~A}{\sqrt{5}}~~\to~~u_i(I) = 3,74x10^{-3}~A
$$

Siendo el error tipo B:

$$
u_j(I) = \sqrt{Var(x)}~~~~~~~Var(x):~varianza = \int_{a}^b x^2f(x)dx
$$

Asumiendo que la medicion tiene una distribucion uniforme continua:

$$
Var(x) = \frac{a^2}{3}~~~~~~a: Precision
$$

Siendo la precision:

$$
a = \Delta V .err\%~~~~~~\Delta V: valor~de~medicion~~~~err\%: error~total~porcentual
$$

Siendo el error porcentual  total:

$$
err\% = err_{med}\% + err_{cont}\%~~~~~~err_{med}\%: error~de~medicion~~~~err_{cont}\%: error~de~conteo
$$

Donde el error porcentual de conteo es:

$$
err_{cont}\% = \frac{total~de~conteos}{valor~de~cuentas~del~instrumento}~.~100\%
$$

$$
err_{cont}(2~conteos)\% = \frac{2}{1000}~.100\%~~\to~~err_{cont}(2~conteos)\% = 0,2\%
$$

$$
err\% = 0,05\% + 0,2\%~~\to~~err\% = 0,25\%
$$

$$
a = \pm~10~A~.~0,25\%~~\to~~a = \pm~0,025~A
$$

Calculando el error tipo B de I:

$$
u_j(I) = \sqrt{\frac{a^2}{3}} = \frac{a}{\sqrt{3}}~~\to~~u_j(I) = \frac{0,025~A}{\sqrt{3}}~~\to~~u_j(I) = 0,0144~A
$$

Calculando la incertidumbre combinada de I:

$$
u_c(I) = \sqrt{u_i^2+u_j^2}~~\to~~u_c(I) = \sqrt{(3,74x10^{-3}~A)^2 + (0,0144~A)^2}~~\to~~u_c(I) = 14,877x10^{-3}~A
$$

Debido a que la incertidumbre tipo B es significativa frente a la tipo A, la distribucion de $u_c(I)$ no se ajusta a una distribucion normal. Es por este motivo que se procede a calcular el valor del factor de cobertura K para un intervalo de confianza del 95,45% con la siguiente tabla:




$$
k_{95,45} = \frac{u_i(I)}{u_j(I)} = \frac{3,74x10^{-3}~A}{0,0144~A} = 0,2597~~\to~~k_{95,45} = 1,75
$$

Siendo la incertidumbre expandida:

$$
U(I) = k_{expansion} . u_c(I) = 1,75~.~14,877x10^{-3}~A~~\to~~U(I) = 26,035x10^{-3}~A
$$

Por lo que el valor de la corriente medida con un grado de confianza de 95% es:

$$
\hat{I} = (10~;\pm~26,035x10^{-3})~A
$$

Siendo el error:

$$
Error = |I_{real} - I_{med}| = |10~A - 10,008~A|~~\to~~Error = 0,008x10^{-3}~A
$$
