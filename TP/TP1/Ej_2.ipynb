# Ejercicio 2

Considere cinco conjuntos independientes de observaciones simultneas de las tres magnitudes de entrada V, I, y  se obtienen en condiciones similares.



a) Cules son las circunstancias en las que debe observarse la correlacin entre las magnitudes de entrada. Un coeficiente de correlacin igual a 0 (cero) indica que la correlacin es alta o baja?

b) Calcule la incertidumbre combinada de las cantidades R, X y Z.

### Parte a

Debe tenerse en cuenta la correlacion entre las magnitudes de entrada cuando estas existan y sean de un valor significativo. Un coeficiente de correlacion igual a cero indica una correlacion lineal nula.

### Parte b

Evaluacion tipo A de V, I y $\Omega$:

$$
u_i = \frac{S}{\sqrt{N}}~~~~~~~S: desviacion~estandard~~~~N: cantidad~de~mediciones
$$

$$
u_i(V) = \frac{0,0032~V}{\sqrt{5}}~~\to~~u_i(V) = 0,00143~V
$$

$$
u_i(I) = \frac{0,0095~mA}{\sqrt{5}}~~\to~~u_i(I) = 0,00425~mA
$$

$$
u_i(\phi) = \frac{0,00075~\phi}{\sqrt{5}}~~\to~~u_i(\phi) = 0,000335
$$

Incertidumbre combinada de V, I y $\Omega$:

$$
u_c = \sqrt{u_i^2+u_j^2}
$$

$$
u_c(V) = \sqrt{(0,00143~V)^2 + (0~V)^2}~~\to~~u_c(V) = 0,00143~V
$$

$$
u_c(I) = \sqrt{(0,00425~mA)^2 + (0~mA)^2}~~\to~~u_c(I) = 0,00425~mA
$$

$$
u_c(\phi) = \sqrt{(0,000335)^2 + (0)^2}~~\to~~u_c(\phi) = 0,000335
$$

Calculando las medias:

$$
\bar{Z} = \frac{\bar{V}}{\bar{I}} = \frac{4,9990~V}{19,6610~mA} = 254,2597019~\Omega
$$

$$
\bar{R} = \bar{Z}.cos(\phi) = 254,26~\Omega.cos(1,04446~\phi) = 127,7321699~\Omega
$$

$$
\bar{X} = \bar{Z}.sen(\phi) = 254,26~\Omega.sen(1,04446~\phi) = 219,8465119~\Omega
$$

Siendo el cuadrado de la incertidumbre combinada:

$$
u_{c}^2(y) = \sum_{i=1}^{N} (\frac{\partial f}{\partial x_i})^2~.~u^2(x_i)~+~2\sum_{i=1}^{N-1}\sum_{j=i+1}^{N}\frac{\partial f}{\partial x_i}\frac{\partial f}{\partial x_j}~.~u(x_i, x_j)
$$

Calculando la incertidumbre combinada de $Z$

$$
u_{c}^2(Z) = (\frac{1}{\bar{I}})^2~.~u_c^2(V) + (\frac{-\bar{V}}{\bar{I}^2})^2~.~u_c^2(I) + 0~.~u_c^2(\phi) + 2~.~(\frac{1}{\bar{I}})~.~(\frac{-\bar{V}}{\bar{I}^2})~.~u_c(V)~.~u_c(I)~.~r(\bar{V},\bar{I})
$$
$$
+ 2~.~(\frac{-\bar{V}}{\bar{I}^2})~.~0~.~u_c(V)~.~u_c(\phi)~.~r(\bar{V},\bar{\phi}) + 2~.~(\frac{1}{\bar{I}})~.~0~.~u_c(V)~.~u_c(\phi)~.~r(\bar{V},\bar{\phi})
$$

$$
u_{c}^2(Z) = (\frac{1}{\bar{I}})^2~.~u_c^2(V) + (\frac{-\bar{V}}{\bar{I}^2})^2~.~u_c^2(I) - 2~.~(\frac{1}{\bar{I}})~.~(\frac{\bar{V}}{\bar{I}^2})~.~u_c(V)~.~u_c(I)~.~r(\bar{V},\bar{I})
$$

Remplazando por los siguientes valores:

$$
\bar{V} = 4,9990~V~~~~\bar{I} = 19,6610~mA~~~~u_c(V) = 0,00143~V~~~~u_c(I) = 0,00425~mA~~~~r(\bar{V}, \bar{I}) = -0,36
$$

$$
u_c^2(Z) = (\frac{1}{19,6610~mA})^2~.~(0,00143~V)^2 + (\frac{4,9990~V}{(19,6610~mA)^2})^2~.~(0,00425~mA)^2
$$
$$
+ 2~\frac{1}{19,6610~mA}~.~\frac{4,9990~V}{(19,6610~mA)^2}~.~0,00143~V~.~0,00425~mA~.~0,36
$$

$$
u_c^2(Z) = 5,29\times10^{-3}~\Omega^2~+~3,02\times10^{-3}~\Omega^2~+~2,88\times10^{-3}~\Omega^2
$$

$$
u_c^2(Z) = 11,19\times10^{-3}~\Omega^2~~\to~~u_c(Z) = 3,345\times10^{-3}~\Omega
$$

Debido a que $u_c(Z)$ es igual a la convolucion de 3 distribuciones distintas, podemos asumir que su distribucion es normal. Por lo tanto el factor de expansion k es igual a 2 y su incertidumbre expandida sera:

$$
u(Z) = k_{expansion}~.~u_c^2(Z) = 2~.~3,345\times10^{-3}~\Omega~~\to~~u(Z) = 6,69\times10^{-3}~\Omega
$$

Siendo el valor de la medicion de $\hat{Z}$ para un intervalo de confianza del 95%:

$$
\hat{Z} = (254,2597~;\pm~0,0067)~\Omega
$$

Calculando la incertidumbre combinada de $R$:

$$
u_c^2(R) = (\frac{1}{\bar{I}})^2~.~u_c^2(V)~.~cos(\phi) + (\frac{-\bar{V}}{\bar{I}^2})^2~.~u_c^2(I)~.~cos(\phi) + \frac{\bar{V}}{\bar{I}}~.~(-sen(\phi))^2~.~u_c^2(\phi) + 2~.~(\frac{1}{\bar{I}})~.~(\frac{-\bar{V}}{\bar{I}^2})~.~u_c(V)~.~u_c(I)~.~r(\bar{V},\bar{I})
$$
$$
+ 2~.~(\frac{-\bar{V}}{\bar{I}^2})~.~(-sen(\phi))~.~u_c(I)~.~u_c(\phi)~.~r(\bar{I},\bar{\phi}) + 2~.~(\frac{1}{\bar{I}})~.~(-sen(\phi))~.~u_c(V)~.~u_c(\phi)~.~r(\bar{V},\bar{\phi})
$$

Remplazando por los siguientes valores:

$$
\bar{V} = 4,9990~V~~~~\bar{I} = 19,6610~mA~~~~\bar{\phi} = 1,0446~~~~u_c(V) = 0,00143~V~~~~u_c(I) = 0,00425~mA~~~~u_c(\phi) = 0,000335
$$
$$
r(\bar{V}, \bar{I}) = -0,36~~~~r(\bar{V}, \bar{\phi}) = 0,86~~~~r(\bar{I}, \bar{\phi}) = -0,65

$$

$$
$$
+ \frac{4,9990~V}{19,6610~mA}~.~(-sen(1,0446))^2~.~(0,000335)^2 + 2~\frac{1}{19,6610~mA}~.~\frac{4,9990~V}{(19,6610~mA)^2}~.~0,00143~V~.~0,00425~mA~.~0,36
$$
$$
+ 2~\frac{4,9990~V}{(19,6610~mA)^2}~.~sen(1,0446)~.~0,000335~.~0,00425~mA~.~0,65+ 2~\frac{1}{19,6610~mA}~.~sen(1,0446)~.~0,0557~V~.~0,000335~.~0,86
$$

$$
u_c^2(R) = 2,657\times10^{-3}~\Omega^2 + 1,517\times10^{-3}~\Omega^2 + 2,133\times10^{-3}~\Omega^2 + 2,878\times10^{-3}~\Omega^2 + 2,069\times10^{-3}~\Omega^2 + 1,411\times10^{-3}~\Omega^2
$$

$$
u_c^2(R) = 12,665\times10^{-3}~\Omega^2~~\to~~u_c(R) = 3,559\times10^{-3}~\Omega
$$

$$
K_{expansion} = 2~~\to~~u(R) = u_c(R)~.~k_{espansion} = 7,118\times10^{-3}~\Omega
$$

Siendo el valor de la medicion de $\hat{R}$ para un intervalo de confianza del 95%:

$$

$$

Calculando la incertidumbre combinada de $X$:

$$

$$
$$
+ 2~.~(\frac{-\bar{V}}{\bar{I}^2})~.~cos(\phi)~.~u_c(I)~.~u_c(\phi)~.~r(\bar{I},\bar{\phi}) + 2~.~(\frac{1}{\bar{I}})~.~cos(\phi)~.~u_c(V)~.~u_c(\phi)~.~r(\bar{V},\bar{\phi})
